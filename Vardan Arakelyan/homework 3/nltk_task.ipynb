```python
import nltk
import re
import pandas as pd
```

```python
df = pd.read_csv('Coachella-2015-2-DFE.csv', encoding ='latin-1')
df.head()
```

```python
# task 1 (hashtags)
reg_for_hashtags = r'#\w+\b'
hashtags = []

for text in df['text']:
    sol_hashtags = re.findall(reg_for_hashtags, text)
    hashtags.append(sol_hashtags)

df['hashtags'] = hashtags
df.head()
```

```python
# task 2 (emails)
reg_for_emails = r'[\w.-]+@\w+\.[A-Za-z]{2,}'
emails = []

for text in df['text']:
    sol_emails = re.findall(reg_for_emails, text)
    if sol_emails:
        emails.append(sol_emails)
    else:
        emails.append('')

df['emails'] = emails
df.head()
```

```python
# task 3

reg_for_username = r'@\w+\b'
reg_for_links = r'\w{4,5}://\S+'
texts = []

for text in df['text']:
    sol_username = re.sub(reg_for_username, '', text)
    sol_links = re.sub(reg_for_links, '',sol_username)
    sol_emails = re.sub(reg_for_emails, '',sol_links)
    sol_hashtags = re.sub(reg_for_hashtags, '',sol_emails)
    texts.append(sol_hashtags.lower())

df['texts'] = texts
df.head()
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')

no_stops = []
for text in df['texts']:
    tokens = nltk.word_tokenize(text)
    stop_words = nltk.corpus.stopwords.words('english')
    filtered_tokens = [token for token in tokens if token not in stop_words]
    final_text = ' '.join(filtered_tokens)
    no_stops.append(final_text)
df['texts'] = no_stops
df.head()
```

```python
reg_for_digits = r'\d+'

no_digits = []
for text in df['texts']:
    sol_digits = re.sub(reg_for_digits, '',text)
    no_digits.append(sol_digits)
df['texts'] = no_digits
df.head()
```

```python
reg_for_chars = r'[^a-zA-Z0-9\s]'
no_chars = []
for text in df['texts']:
    sol_chars = re.sub(reg_for_chars, '',text)
    no_chars.append(sol_chars)
df['texts'] = no_chars
df.head()
```
